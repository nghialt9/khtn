### `1. Hướng dẫn`

- `Đoạn code sau cài đặt BFS theo phương pháp đơn giản.`

```python
# Định nghĩa đồ thị theo dạng danh sách kề
graph = { 
    'A': ['B', 'C'], 
    'B': ['D', 'E'], 
    'C': ['F'], 
    'D': [], 
    'E': ['F'], 
    'F': [] 
} 

# Danh sách để theo dõi các đỉnh đã được thăm
visited = [] 

# Hàng đợi (Queue) để quản lý các đỉnh đang chờ được thăm
queue = []     

def bfs(visited, graph, start, end): 
    # Thêm đỉnh xuất phát vào hàng đợi
    queue.append(start) 

    while queue: 
        # Lấy đỉnh đầu tiên từ hàng đợi
        s = queue.pop(0)  

        # In ra đỉnh hiện tại
        print(s, end=" ")  

        # Kiểm tra nếu đỉnh hiện tại là đích
        if s == end: 
            return 

        # Đánh dấu đỉnh này là đã thăm
        visited.append(s) 

        # Thêm các đỉnh kề vào hàng đợi nếu chưa thăm
        for neighbour in graph[s]: 
            if neighbour not in visited: 
                queue.append(neighbour) 

# Thực thi thuật toán BFS
bfs(visited, graph, 'A', 'F')
```

- `Đoạn code tiếp theo vẫn cài đặt BFS nhưng có xuất ra đường đi`

```python
# Định nghĩa đồ thị theo dạng danh sách kề
graph = { 
    '1': ['2', '3', '4'], 
    '2': ['5', '6'], 
    '5': ['9', '10'], 
    '4': ['7', '8'], 
    '7': ['11', '12'] 
} 

def bfs(graph, start, end): 
    # Khởi tạo hàng đợi với đường đi từ đỉnh xuất phát
    queue = [[start]] 
    visited = []

    while queue: 
        # Lấy đường đi đầu tiên từ hàng đợi
        path = queue.pop(0) 
        # Lấy đỉnh cuối cùng từ đường đi
        node = path[-1] 

        # Nếu đã đến đỉnh đích, trả về đường đi
        if node == end: 
            return path 

        # Đánh dấu đỉnh hiện tại là đã thăm
        visited.append(node)

        # Thêm các đường đi mới với các đỉnh kề chưa thăm vào hàng đợi
        for neighbour in graph.get(node, []): 
            if neighbour not in visited: 
                new_path = list(path) 
                new_path.append(neighbour) 
                queue.append(new_path) 

# Thực thi thuật toán BFS và in ra đường đi tìm được
print(bfs(graph, '1', '11'))
```

+ `2.2 Dựa vào đoạn code mẫu về BFS. Sinh viên cài đặt thuật toán DFS và cho biết
kết quả thực thi DFS với các đồ thị như phần 2.1`

```python
# Code DFS
def dfs(graph, start, end):
    # Khởi tạo ngăn xếp với đường đi từ đỉnh xuất phát
    stack = [[start]]
    visited = []

    while stack:
        # Lấy đường đi đầu tiên từ ngăn xếp
        path = stack.pop()
        # Lấy đỉnh cuối cùng từ đường đi
        node = path[-1]

        # Nếu đã đến đỉnh đích, trả về đường đi
        if node == end:
            return path

        # Nếu đỉnh chưa được thăm, tiến hành thăm
        if node not in visited:
            visited.append(node)

            # Thêm các đường đi mới với các đỉnh kề vào ngăn xếp
            for neighbour in graph.get(node, []):
                new_path = list(path)
                new_path.append(neighbour)
                stack.append(new_path)
```

```python
from collections import deque, defaultdict

def is_bipartite(graph, start):
    # Khởi tạo các tập X và Y
    X = set()
    Y = set()
    color = {}  # Lưu trữ màu của từng đỉnh

    queue = deque([start])
    color[start] = 0  # Bắt đầu với màu 0

    while queue:
        node = queue.popleft()
        current_color = color[node]
        
        # Xác định tập X hoặc Y dựa trên màu sắc
        if current_color == 0:
            X.add(node)
        else:
            Y.add(node)

        # Xét các đỉnh kề
        for neighbor in graph[node]:
            if neighbor not in color:
                # Gán màu cho đỉnh kề
                color[neighbor] = 1 - current_color
                queue.append(neighbor)
            elif color[neighbor] == color[node]:
                # Nếu đỉnh kề có màu giống như đỉnh hiện tại, đồ thị không phải là lưỡng phân
                return False, None, None

    return True, X, Y

def main():
    # Định nghĩa đồ thị dưới dạng danh sách kề
    graphY = {
    1: [2, 3, 4],
    2: [1, 5],
    3: [1, 5],
    4: [1, 5],
    5: [2, 3, 4],
    }

    graphN = {
        1: [2, 14, 6],
        2: [1, 13, 3],
        3: [2, 10, 4],
        4: [3, 7,5],
        5: [4, 8, 6],
        6: [1, 5, 11],
        7: [4, 8, 10],
        8: [5, 7, 9],
        9: [8, 10, 12],
        10: [3, 7, 9],
        11: [6, 12, 14],
        12: [9, 11, 13],
        13: [2, 12, 14],
        14: [1, 11, 13]
    }

    # Kiểm tra đồ thị từ đỉnh bắt đầu (ví dụ: đỉnh 1)
    start_vertex = 1
    bipartite, X, Y = is_bipartite(graphY, start_vertex)

    if bipartite:
        print("Đồ thị là đồ thị lưỡng phân.")
        print("Tập X (đỉnh trái):", X)
        print("Tập Y (đỉnh phải):", Y)
    else:
        print("Đồ thị không phải là đồ thị lưỡng phân.")

if __name__ == "__main__":
    main()
```

```python
from collections import deque

def is_bipartite(graph):
    color = {}
    
    for node in graph:
        if node not in color:
            color[node] = 0  # Khởi tạo màu sắc ban đầu cho node
            queue = deque([node])
            
            while queue:
                current = queue.popleft()
                
                for neighbor in graph[current]:
                    if neighbor not in color:
                        color[neighbor] = 1 - color[current]  # Gán màu đối lập cho neighbor
                        queue.append(neighbor)
                    elif color[neighbor] == color[current]:
                        return False  # Nếu có cạnh giữa hai đỉnh cùng màu, không phải lưỡng phân
    return True

# Định nghĩa đồ thị
graphY = {
    1: [2, 3, 4],
    2: [1, 5],
    3: [1, 5],
    4: [1, 5],
    5: [2, 3, 4],
}

graphN = {
    1: [2, 14, 6],
    2: [1, 13, 3],
    3: [2, 10, 4],
    4: [3, 7,5],
    5: [4, 8, 6],
    6: [1, 5, 11],
    7: [4, 8, 10],
    8: [5, 7, 9],
    9: [8, 10, 12],
    10: [3, 7, 9],
    11: [6, 12, 14],
    12: [9, 11, 13],
    13: [2, 12, 14],
    14: [1, 11, 13]
}

# Kiểm tra đồ thị
if is_bipartite(graphN):
    print("Đồ thị là lưỡng phân.")
else:
    print("Đồ thị không phải là lưỡng phân.")
```
