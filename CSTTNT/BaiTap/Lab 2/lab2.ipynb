### ***Code các hàm***

```python
# A* Algorithm with node expansion count and comments

from collections import deque

class Graph:
    # Initialize the graph with adjacency list and heuristic
    def __init__(self, adjacency_list, heuristic):
        self.adjacency_list = adjacency_list  # Store the adjacency list representing the graph
        self.h = heuristic  # Store the heuristic values for each node
        self.expanded_nodes = 0  # Initialize the counter for expanded nodes

    # Get the neighbors of a node
    def get_neighbors(self, v):
        return self.adjacency_list[v]  # Return the list of neighbors and edge weights for node v

    # A* search algorithm
    def a_star_algorithm(self, start_node, stop_node):
        open_list = set([start_node])  # Initialize the open list with the start node
        closed_list = set([])  # Initialize the closed list as empty

        g = {}  # Dictionary to store the cost from the start node to each node
        g[start_node] = 0  # Cost from the start node to itself is 0

        parents = {}  # Dictionary to store the parent of each node
        parents[start_node] = start_node  # Start node is its own parent

        while len(open_list) > 0:  # Continue until there are no more nodes to explore
            n = None  # Initialize the current node

            # Find the node with the lowest f value (g + h)
            for v in open_list:
                if n == None or g[v] + self.h[v] < g[n] + self.h[n]:
                    n = v

            if n == None:  # If no node is found, there is no path
                print('Path does not exist!')
                return None

            if n == stop_node:  # If the goal node is reached, reconstruct the path
                reconst_path = []

                while parents[n] != n:  # Trace back from the goal to the start node
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start_node)
                reconst_path.reverse()  # Reverse the path to get the correct order

                print(f'Path found: {reconst_path}')
                print(f'Nodes expanded: {self.expanded_nodes}')
                return reconst_path

            # Explore the neighbors of the current node
            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:  # If neighbor is not explored
                    open_list.add(m)  # Add to the open list
                    parents[m] = n  # Set the current node as the parent
                    g[m] = g[n] + weight  # Update the cost to reach this neighbor
                    self.expanded_nodes += 1  # Increment the count of expanded nodes
                else:  # If the neighbor is already explored, check if a cheaper path exists
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight  # Update the cost if a cheaper path is found
                        parents[m] = n  # Update the parent
                        self.expanded_nodes += 1

                        if m in closed_list:  # Move the neighbor back to the open list if necessary
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)  # Move the current node from open to closed list
            closed_list.add(n)

        print('Path does not exist!')
        return None

# Uniform Cost Search (UCS) Implementation with comments
class UCS(Graph):
    def ucs_algorithm(self, start_node, stop_node):
        open_list = set([start_node])  # Same as A*, but without using the heuristic
        closed_list = set([])

        g = {}
        g[start_node] = 0

        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            # Select the node with the lowest cost (g value)
            for v in open_list:
                if n == None or g[v] < g[n]:
                    n = v

            if n == None:
                print('Path does not exist!')
                return None

            if n == stop_node:
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start_node)
                reconst_path.reverse()

                print(f'Path found: {reconst_path}')
                print(f'Nodes expanded: {self.expanded_nodes}')
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                    self.expanded_nodes += 1
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        self.expanded_nodes += 1

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)
            closed_list.add(n)

        print('Path does not exist!')
        return None

# Greedy Search Implementation with comments
class Greedy(Graph):
    def greedy_algorithm(self, start_node, stop_node):
        open_list = set([start_node])  # Same as A*, but only using the heuristic
        closed_list = set([])

        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            # Select the node with the lowest heuristic value (h)
            for v in open_list:
                if n == None or self.h[v] <= self.h[n]:
                    n = v

            if n == None:
                print('Path does not exist!')
                return None

            if n == stop_node:
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start_node)
                reconst_path.reverse()

                print(f'Path found: {reconst_path}')
                print(f'Nodes expanded: {self.expanded_nodes}')
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    self.expanded_nodes += 1

            open_list.remove(n)
            closed_list.add(n)
        
        print('Path does not exist!')
        return None
```
